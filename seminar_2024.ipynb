```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

df=pd.read_excel(r'/content/drive/MyDrive/python_bank/machine_book/專題2024/20220121.xlsx')
df=df.fillna(method='ffill')
df.tail()
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[0:-1, 3:]
print(x)

y=df.iloc[1:, 2]
print(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)  #先訓練80%，再用訓練的結果測20%
# 觀看數據集、訓練與測試集的類別比例
df_count = pd.concat([y.value_counts(), y_train.value_counts(), y_test.value_counts()], axis=1)
df_count.columns = ['y', 'y_train', 'y_test']
df_count
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
clfr = RandomForestClassifier(max_depth=3, n_jobs=-1)
clfr.fit(x, y)   #開始計算
y_pred = clfr.predict(x)
#print(classification_report(y, y_pred))
```

```python
"""
importances = clfr.feature_importances_
std = np.std([t.feature_importances_ for t in clfr.estimators_], axis=0)
idx = np.argsort(importances)[::-1]

plt.title("Feature importances")
plt.bar(range(x.shape[1]), importances[idx], yerr=std[idx], align="center")
plt.xticks(range(x.shape[1]), labels=x.columns[idx])
plt.xlim([-1, x.shape[1]])
plt.ylim([0, 0.6])
"""
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
# 建立特徵選取器，門檻值預設為重要性的平均值
selectorr = SelectFromModel(clfr)
selectorr.fit(x, y)
print('門檻值 =', selectorr.threshold_)
print('特徵遮罩：', selectorr.get_support())

# 選出新特徵，重新訓練隨機森林
x_new = selectorr.transform(x)
x_new=pd.DataFrame(x_new)
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42).split(x, y)
score_lst = []  # 紀錄 k 次交叉驗證的正確率

for k, (i_train, i_valid) in enumerate(kfold):
  clfr.fit(x_new.iloc[i_train, :], y.iloc[i_train])

  y_pred_test = clfr.predict(x_new.iloc[i_valid, :])

  score=accuracy_score(y.iloc[i_valid], y_pred_test)
  print('precision=',score)
  score_lst.append(score)
  print('%2d-Fold: Acc=%.3f' % (k+1, score))

print('\n10-fold CV accuracy = %.3f, std = %.3f' % (np.mean(score_lst), np.std(score_lst)))
print(np.mean(score_lst)/np.std(score_lst))
```

## **Chapter 2_Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
clfl = LogisticRegression()
clfl.fit(x, y)
y_pred = clfl.predict(x)
print(classification_report(y, y_pred))
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
# 建立特徵選取器，門檻值預設為重要性的平均值
selectorl = SelectFromModel(clfl)
selectorl.fit(x, y)
print('門檻值 =', selectorl.threshold_)
print('特徵遮罩：', selectorl.get_support())

# 選出新特徵，重新訓練隨機森林
x_new = selectorl.transform(x)
x_new=pd.DataFrame(x_new)
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42).split(x, y)
score_lst = []  # 紀錄 k 次交叉驗證的正確率

for k, (i_train, i_valid) in enumerate(kfold):
  clfl.fit(x_new.iloc[i_train, :], y.iloc[i_train])

  y_pred_test = clfl.predict(x_new.iloc[i_valid, :])

  score=accuracy_score(y.iloc[i_valid], y_pred_test)
  print('precision=',score)
  score_lst.append(score)
  print('%2d-Fold: Acc=%.3f' % (k+1, score))

print('\n10-fold CV accuracy = %.3f, std = %.3f' % (np.mean(score_lst), np.std(score_lst)))
print(np.mean(score_lst)/np.std(score_lst))
```
