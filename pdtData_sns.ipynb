```python
import pandas as pd
url='https://raw.githubusercontent.com/bluecityisu/class2022/main/pdtData_Sales.csv'
data = pd.read_csv(url)
```

```python
from matplotlib import pyplot as plt
```

```python
t=data.iloc[:,0]
qty=data.iloc[:,2]
#category=data.iloc[:,8]
#season=data.iloc[:,5]
sales=data.iloc[:,6]
area=data.iloc[:,4]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,8))

axs[0].plot(t,sales,'k')
axs[0].set_title("Sales", fontsize='10')
axs[0].set(ylabel='sales')

axs[1].plot(t,qty,'b')
axs[1].set_title("Quantity",fontsize='10')
axs[1].set(ylabel='Qty')

plt.subplots_adjust(top=0.4, wspace=0.2, hspace=0.4)
plt.show()
```

```python
import seaborn as sns
features = sns.pairplot(data, vars=['qty', 'sales'], hue='area')
```
